```python
#check the contents of "GRADE" to create bar count
df['GRADE'].unique()
```

```python
#assignment of color for bar count
color=['EBD69D','#F08030',  
                    '#6890F0',  
                    '#A8B820',  
                    '#A8A878', 
                    '#A040A0', 
                    '#F8D030',  
                    '#E0C068',  
                    '#EE99AC',  
                    '#C03028',  
                    '#F85888',  
                    '#B8A038',  
                    '#705898', 
                    '#98D8D8', 
                    '#7038F8']
```

```python
#chart 3: Bar Count

sns.countplot(x='GRADE',
             palette=color,
             data=df)
plt.xticks(rotation=-45) 
plt.title('Frequency by level of grade in DC properties')
```
