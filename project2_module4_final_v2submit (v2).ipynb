```python
#chart 4: box plot 1
df4=df.loc[:,['WARD','PRICE']]
```

```python
plt.subplots(figsize=(10,10))
sns.boxplot(x='WARD',y='PRICE', data=df4)
```

```python
#4: boxplot 2 - delete Ward2, 3 and 8 with outliers
df5=df4[(df4['WARD'] != 'Ward 2') & (df4['WARD'] != 'Ward 3') & (df4['WARD'] != 'Ward 8')] 
```

```python
plt.subplots(figsize=(10,8))
sns.boxplot(x='WARD',y='PRICE', data=df5)
```

```python
#4 boxplot 2: delete outliers using interquartile range method

q1=df4['PRICE'].quantile(.25)
q3=df4['PRICE'].quantile(.75)

iqr=q3-q1

toprange=q3+iqr
botrange=q1-iqr

df6=df4.copy()

df6=df6.drop(df6[df6['PRICE']>toprange].index)
df6=df6.drop(df6[df6['PRICE']<botrange].index)

print(df6.shape)
df6.head()
```

```python
plt.subplots(figsize=(10,8))
sns.boxplot(x='WARD',y='PRICE', data=df6)
```
